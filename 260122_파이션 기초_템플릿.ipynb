# 4장 프로그램의 선택 (조건문)

### 양자택일: if ~ else 문
- 조건이 참이면 if 블록을, 거짓이면 else 블록을 실행합니다

```python
if money >= 3000:
    print("택시를 탄다")
else:
    print("걸어간다") # 조건이 거짓일 때 실행
```

### 다중 선택: if ~ elif ~ else
- 조건이 여러 개일 때 사용합니다. elif(else if의 줄임말)는 개수 제한 없이 쓸 수 있습니다. 
- 위에서부터 차례대로 검사하며, 하나라도 참이면 나머지는 건너뜁니다.

```python
score = 85

if score >= 90:
    print("A학점")
elif score >= 80:
    print("B학점") # 여기서 참이 되므로 실행하고 if문 종료
elif score >= 70:
    print("C학점")
else:
    print("F학점")
```

###  중첩 조건문
- if문 안에 또 다른 if문을 넣을 수 있습니다

```python
age = 20
card = True

if age >= 19:
    if card:
        print("결제 진행")
    else:
        print("카드가 필요합니다")
else:
    print("미성년자는 구매 불가")
```

### 조건에 따른 반복: while 문
- 조건식이 참(True)인 동안 계속 반복합니다. 반복 횟수를 정확히 모를 때 유용합니다.
- 조건이 영원히 True라면 프로그램이 멈추지 않는 **무한 루프(Infinite Loop**)에 빠집니다
이때는 `Ctrl + C`를 눌러 강제 종료하세요.

```python
count = 0
while count < 3:
    print(f"현재 카운트: {count}")
    count += 1 # 이 줄이 없으면 count가 계속 0이라 무한 루프 발생!
```

### 흐름 제어: break와 continue
| 제어문          | 기능                                | 비유              |
| ------------ | --------------------------------- | --------------- |
| **break**    | 반복문을 **즉시 탈출**합니다.                | 게임 종료!          |
| **continue** | 아래 코드를 실행하지 않고 **다음 반복으로 건너뜁니다.** | 이번 판 무효, 다시 시작! |

```python
# 1부터 10까지 출력하되, 5가 되면 멈춤
for i in range(1, 11):
    if i == 5:
        break
    print(i, end=" ") # 출력: 1 2 3 4
```

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i, end=" ") # 출력(5제외): 1 2 3 4 6 7 8 9 10
```

### 중첩 반복문 (이중 for문)
- 반복문 안에 반복문이 있는 형태입니다. 시계의 시침과 분침을 생각하면 쉽습니다.
- 바깥쪽이 한 번 돌 때, 안쪽은 전체를 돕니다.

```python
# 구구단
for i in range(2, 10):       # 2단부터 9단까지
    print(f"=== {i}단 ===")
    for j in range(1, 10):   # 곱해지는 수 1~9
        print(f"{i} x {j} = {i*j}")
```

```python
# 숫자 맞히기 게임
import random

# 1. 초기 설정
secret_number = random.randint(1, 100)
attempts = 0          # 현재 시도 횟수
max_attempts = 10     # 최대 허용 횟수
is_win = False        # 승리 여부 확인용 플래그

print("=== 숫자 맞히기 게임 (10회 제한) ===")
print(f"1부터 100 사이의 숫자 하나를 생각했습니다. {max_attempts}번 안에 맞혀보세요!")
print("(게임을 중단하려면 'QUIT'를 입력하세요.)") # 안내 문구 추가

# 2. 반복문 시작
while attempts < max_attempts:
    # 사용자 입력을 먼저 문자열로 받음
    user_input = input(f"[{attempts + 1}회차] 숫자를 입력하세요: ").strip()
    
    # 2-1. 종료 조건 확인 (대소문자 구분 없이 처리)
    if user_input.upper() == "QUIT":
        print("\n게임을 종료합니다. 다음에 또 만나요!")
        break

    try:
        # 입력을 정수로 변환
        guess = int(user_input)
        attempts += 1
        
        # 3. 비교 결과 판정
        if guess < secret_number:
            print("UP! 더 큰 숫자를 입력해보세요.")
        elif guess > secret_number:
            print("DOWN! 더 작은 숫자를 입력해보세요.")
        else:
            print(f"축하합니다! {attempts}번 만에 숫자를 맞히셨습니다! 🎉")
            is_win = True
            break  # 정답을 맞혔으므로 반복문 즉시 탈출
            
    except ValueError:
        print("숫자나 'QUIT'만 입력할 수 있습니다. 다시 시도하세요.")

# 4. 반복문 종료 후 결과 처리
# 정답을 맞히지 못했거나 'QUIT'로 나가지 않은 경우 (시도 횟수 초과 시)
if not is_win and attempts == max_attempts:
    print("\n" + "="*30)
    print("아쉽네요! 모든 기회를 다 사용하셨습니다.")
    print(f"정답은 [ {secret_number} ]였습니다. 다음에 다시 도전하세요!")
    print("="*30)
```

#### Type 3. 선택형 (조건에 따라 다르게)

```python
scores = [80, 50, 90, 40]
# 60점 이상이면 '합격', 아니면 '불합격'
results = ["합격" if s >= 60 else "불합격" for s in scores]
print(results) # ['합격', '불합격', '합격', '불합격']
```

### 복사의 함정: 참조 복사
#### 가변(Mutable)
- 생성 후에도 내용을 바꿀 수 있습니다. (리스트, 딕셔너리, 집합)
#### 불변(Immutable)
- 생성 후에는 내용을 바꿀 수 없습니다. (정수, 실수, 문자열, 튜플)
#### 리스트 같은 가변 객체를 = 등호로 다른 변수에 넣으면, 
- 데이터 자체가 복사되는 것이 아니라 '주소표(참조)'만 복사됩니다.
- 즉, 두 변수가 하나의 데이터를 같이 바라보게 됩니다.

```python
# [참조 복사]
a = [1, 2, 3]
b = a  # 데이터 복사가 아니라, a가 가리키는 주소를 b도 가리키게 됨

b[0] = 100 # b를 수정했는데...
print(a)   # [100, 2, 3] -> 헉! a도 같이 바뀌어 버립니다.
```

### 원본(a)을 보호하면서 복사본(b)만 수정하고 싶다면, 
- copy() 메서드나 
- 슬라이싱 [:]을 사용하여 새로운 주소에 데이터를 복제해야 합니다.

```python
# [얕은 복사]
a = [1, 2, 3]
b = a.copy() # 또는 b = a[:]
b[0] = 100
print(a) # [1, 2, 3] -> 원본은 안전합니다.
print(b) # [100, 2, 3] -> 복사본만 바뀝니다.
```

### 변수의 범위 (Scope)
- 지역 변수(Local):** 함수 **안**에서 만든 변수. 함수가 끝나면 사라집니다. (밖에서 못 씀) 
- **전역 변수(Global):** 함수 **밖**에서 만든 변수. 어디서든 쓸 수 있습니다.

```python
global_var = 100 # 전역 변수

def my_func():
    local_var = 20 # 지역 변수
    print(global_var) # 읽기 가능

my_func()
# print(local_var) # 에러! (함수 밖에서는 지역 변수가 안 보임)
```

### 이름 없는 함수: 람다(Lambda)
- 보통 함수는 def로 이름을 짓고 만들지만, 한 번만 쓰고 버릴 간단한 함수는 **람다(Lambda)**로 만듭니다. '익명 함수'라고도 부릅니다.
- **문법:** `lambda 매개변수: 표현식`  
- **특징:** `return` 문 없이도 결과가 자동으로 반환됩니다.

```python
# 일반 함수
def add(a, b):
    return a + b

# 람다 함수 (위와 같은 기능, 이름 없음)
print(add(3, 5)) # 8
print((lambda a,b : a+b)(3,5))
```

### 한꺼번에 처리하기: map() 함수
- 반복문(`for`)을 쓰지 않고도 한 줄로 처리할 수 있어 매우 효율적입니다.

```python
names = ["Kim Python", "Lee Java", "Park C++"]

# 공백 기준으로 자른 후([0]) 성만 가져오기
def split_text(x):
    return x.split()[0]
    
last_names = list(map(split_text, names)) 
print(last_names) # ['Kim', 'Lee', 'Park']


```

# 부록 효율적인 설계도 (클래스와 객체)
- 지금까지는 변수와 함수를 따로 관리했습니다.
- 하지만 게임 캐릭터나 AI 모델처럼 **"속성(변수)"과 "행동(함수)"을 하나로 묶어 관리**해야 할 때가 많습니다. 
- 이것이 바로 **객체지향 프로그래밍(OOP)**입니다.
- 가장 유명한 비유인 **붕어빵 틀**을 기억하세요.
- **클래스 (Class):** 붕어빵 **틀** (설계도)
- **객체/인스턴스 (Object/Instance):** 틀로 찍어낸 **붕어빵** (실체)

```python
class FishBread:
    # 생성자: 붕어빵이 태어날 때 실행됨
    def __init__(self, flavor, price):
        self.flavor = flavor # 맛 (속성)
        self.price = price   # 가격 (속성)

    # 메서드: 붕어빵의 행동
    def show_info(self):
        # 여기서 self는 호출한 녀석(bread1 또는 bread2)이 됩니다.
        print(f"{self.flavor} 붕어빵은 {self.price}원입니다.")

# 객체 생성 (틀로 빵 찍어내기)
bread1 = FishBread("팥", 1000)      # self가 bread1이 되어 '팥'을 저장
bread2 = FishBread("슈크림", 1200)  # self가 bread2가 되어 '슈크림'을 저장

bread1.show_info() # bread1의 정보 출력
bread2.show_info() # bread2의 정보 출력
```

#### **객체 생성 과정의 비밀 (4단계**)
`bread1 = FishBread("팥", 1000)` 코드가 실행될 때, 컴퓨터 내부에서는 다음과 같은 일이 일어납니다.
1. **빈 붕어빵 탄생:** 파이썬이 메모리에 **아무 내용 없는 빈 객체**를 하나 만듭니다.
2. **`self`로 전달:** 이 빈 객체를 `__init__` 함수의 첫 번째 파라미터인 `self`에게 쏙 넣어줍니다. (즉, 지금부터 `self`는 방금 만든 그 빈 붕어빵입니다.)
3. **속성 부여:** `__init__` 함수 안의 `self.flavor = flavor` 코드가 실행되면서, 빈 붕어빵(`self`)에 '팥'이라는 내용물이 채워집니다. 
4. **변수에 저장:** 내용물이 다 채워진 완성된 붕어빵이 `bread1`이라는 변수에 담깁니다.

#### **메서드에서 `self`가 변하는 마법**
객체를 만들고 나서 `bread1.show_info()`를 실행하면 무슨 일이 일어날까요?
- 우리는 `show_info()` 괄호 안에 아무것도 안 넣었지만, 파이썬이 알아서 **주인공**을 넣어줍니다.  
- `bread1.show_info()` ➡ 실제로는 `FishBread.show_info(bread1)`처럼 동작합니다.
- 함수 정의의 `def show_info(self):`에서 `self` 자리에 `bread1`이 들어가는 것이죠.
- 그래서 `print(self.flavor)`는 `print(bread1.flavor)`가 되어 "팥"이 출력되는 것입니다.

### **문제:** `Car` 클래스를 만드세요.
1. 속성: `model`(모델명), `speed`(속도, 초기값 0)
2. 메서드 `accelerate(value)`: 속도가 value만큼 증가
3. 메서드 `brake(value)`: 속도가 value만큼 감소 (0 밑으로는 안 내려감)
4. 차를 한 대 뽑아서 달리고 멈춰보세요.

```python
class Car:
    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self, value):
        self.speed += value
        print(f"{self.model} 가속! 현재 속도: {self.speed}km/h")

    def brake(self, value):
        self.speed -= value
        if self.speed < 0:
            self.speed = 0
        print(f"{self.model} 감속! 현재 속도: {self.speed}km/h")

my_car = Car("테슬라")
my_car.accelerate(50)
my_car.brake(20)
```

### 클래스의 확장: 상속 (Inheritance)
- 이미 만들어진 클래스의 기능을 물려받아 새로운 기능을 추가하는 것입니다.
    (예: '붕어빵' 클래스를 상속받아 '잉어빵' 만들기)

```python
# FishBread의 기능을 물려받음
class PremiumFishBread(FishBread):
    def discount(self):
        self.price -= 100 # 100원 할인 기능 추가

p_bread = PremiumFishBread("황금", 2000)
p_bread.discount()
p_bread.show_info() # 황금 붕어빵은 1900원입니다.
```

```python
# 학생 성적 처리 
class Student:
    """
    개별 학생의 정보를 관리하는 클래스.
    - name: 학생 이름 (문자열)
    - grades: 과목별 점수를 저장하는 딕셔너리, 예) {'Math': 90, 'English': 85}
    """
    def __init__(self, name):
        self.name = name
        self.grades = {}

    def add_grade(self, subject, score):
        """
        특정 과목(subject)에 점수(score)를 추가하거나 갱신한다.
        score는 0~100 사이의 정수여야 한다.
        """
        if not (0 <= score <= 100):
            raise ValueError("점수는 0에서 100 사이의 정수여야 합니다.")
        self.grades[subject] = score

    def remove_grade(self, subject):
        """
        특정 과목(subject)의 성적을 삭제한다. 과목이 없으면 KeyError가 발생한다.
        """
        if subject in self.grades:
            del self.grades[subject]
        else:
            raise KeyError(f"'{subject}' 과목이 존재하지 않습니다.")

    def get_average(self):
        """
        현재 저장된 모든 과목의 평균 점수를 계산하여 반환한다.
        과목이 하나도 없으면 None을 반환한다.
        """
        if not self.grades:
            return None
        return sum(self.grades.values()) / len(self.grades)

    def get_grade_report(self):
        """
        과목별 점수와 평균 점수를 문자열 형태로 반환한다.
        """
        if not self.grades:
            return f"{self.name} 학생의 성적이 없습니다."
        
        report_lines = [f"{self.name} 학생의 성적표:"]
        for subject, score in self.grades.items():
            report_lines.append(f"  - {subject}: {score}점")
        
        avg = self.get_average()
        report_lines.append(f"  - 평균: {avg:.2f}점")
        report_lines.append(f"  - 학점: {self._get_letter_grade(avg)}")
        return "\n".join(report_lines)

    def _get_letter_grade(self, avg_score):
        """
        평균 점수(avg_score)에 따라 A/B/C/D/F 학점을 계산하여 반환한다.
        """
        if avg_score is None:
            return "-"
        if avg_score >= 90:
            return "A"
        elif avg_score >= 80:
            return "B"
        elif avg_score >= 70:
            return "C"
        elif avg_score >= 60:
            return "D"
        else:
            return "F"

    def __str__(self):
        return self.get_grade_report()
```
