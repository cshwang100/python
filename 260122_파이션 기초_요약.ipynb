### 형 변환(Type Casting)
- 자료형을 다른 형태로 강제로 바꾸는 것입니다.
| 함수         | 설명      | 예시                   |
| ---------- | ------- | -------------------- |
| `int(값)`   | 정수로 변환  | `int("10")` → `10`   |
| `float(값)` | 실수로 변환  | `float(5)` → `5.0`   |
| `str(값)`   | 문자열로 변환 | `str(100)` → `"100"` |

```python
# 엑셀이나 웹에서 긁어온 데이터는 보통 문자열 형태
raw_price = "1,500" 
# 쉼표(,)를 제거합니다. "1,500" -> "1500"
clean_str = raw_price.replace(",", "")
# 계산을 위해 정수(int)로 변환합니다.
price = int(clean_str)
# 이제 덧셈 계산이 가능합니다!
print(f"배송비 포함: {price + 3000}원")
```

# 3장 컴퓨터의 계산기 (연산자)

### 산술 연산자
| 연산자  | 의미   | 설명 및 예시                       |
| ---- | ---- | ----------------------------- |
| `+`  | 더하기  | `3 + 2` → `5`                 |
| `-`  | 빼기   | `3 - 2` → `1`                 |
| `*`  | 곱하기  | `3 * 2` → `6`                 |
| `/`  | 나누기  | `4 / 2` → `2.0` (항상 실수 결과)    |
| `//` | 몫    | `10 // 3` → `3` (나눗셈의 정수 부분)  |
| `%`  | 나머지  | `10 % 3` → `1` (짝수/홀수 판별에 유용) |
| `**` | 거듭제곱 | `2 ** 3` → `8` ($2^3$)        |

```python
print(10 // 3) # 3 (10을 3으로 나눈 몫)
print(10 % 3)  # 1 (10을 3으로 나눈 나머지)
```

# 4장 프로그램의 선택 (조건문)
- if 문의 기본 구조와 들여쓰기
- if 문 다음에 오는 실행 코드는 반드시 들여쓰기(Tab 키 또는 스페이스 4칸)를 해야 if`문에 속한 코드로 인식

```python
money = 2000

if money >= 3000:
    print("택시를 탄다") # 들여쓰기 필수!
print("이 줄은 조건과 상관없이 항상 실행됩니다.")
```

### 양자택일: if ~ else 문
- 조건이 참이면 if 블록을, 거짓이면 else 블록을 실행합니다

```python
if money >= 3000:
    print("택시를 탄다")
else:
    print("걸어간다") # 조건이 거짓일 때 실행
```

### 다중 선택: if ~ elif ~ else
- 조건이 여러 개일 때 사용합니다. elif(else if의 줄임말)는 개수 제한 없이 쓸 수 있습니다. 
- 위에서부터 차례대로 검사하며, 하나라도 참이면 나머지는 건너뜁니다.

```python
score = 85

if score >= 90:
    print("A학점")
elif score >= 80:
    print("B학점") # 여기서 참이 되므로 실행하고 if문 종료
elif score >= 70:
    print("C학점")
else:
    print("F학점")
```

### 리스트 메서드(Method)
| **분류**    | **메서드**            | **설명**                                            | **예시 코드**                                                   | **결과 (a 또는 반환값)**                            |
| --------- | ------------------ | ------------------------------------------------- | ----------------------------------------------------------- | -------------------------------------------- |
| **추가**    | `append(x)`        | 리스트의 끝에 항목 `x`를 하나 추가합니다.                         | `a = [1, 2]` `a.append(3)`                 | `[1, 2, 3]`                                  |
|           | `extend(iterable)` | 리스트 끝에 다른 리스트나 반복 가능한 객체의 모든 항목을 추가합니다.           | `a = [1, 2]` `a.extend([3, 4])`            | `[1, 2, 3, 4]`                               |
|           | `insert(i, x)`     | 지정한 인덱스 `i` 위치에 항목 `x`를 삽입합니다.                    | `a = [1, 3]` `a.insert(1, 2)`              | `[1, 2, 3]`                                  |
| **삭제**    | `pop(i)`           | 인덱스 `i`의 항목을 삭제하고 그 값을 반환합니다. (인덱스 생략 시 마지막 항목)   | `a = [1, 2, 3]` `val = a.pop()`            | `a: [1, 2]` `val: 3`        |
|           | `remove(x)`        | 리스트에서 첫 번째로 나오는 항목 `x`를 삭제합니다.                    | `a = [1, 2, 3, 2]` `a.remove(2)`           | `[1, 3, 2]`                                  |
|           | `clear()`          | 리스트의 모든 항목을 삭제하여 빈 리스트로 만듭니다.                     | `a = [1, 2, 3]` `a.clear()`                | `[]`                                         |
| **탐색/확인** | `index(x)`         | 리스트에서 항목 `x`가 있는 첫 번째 인덱스를 반환합니다.                 | `a = ['a', 'b', 'c']` `idx = a.index('b')` | `1` (반환값)                                    |

| **정렬/순서** | `sort()`           | 리스트를 오름차순으로 원본 자체를 정렬합니다. (`reverse=True` 시 내림차순) | `a = [3, 1, 2]` `a.sort()`                 | `[1, 2, 3]`                                  |

### 리스트 컴프리헨션(List Comprehension)
- for 반복문을 한 줄로 압축하여 새로운 리스트를 만듭니다

### Type 1. 기본형 (모든 데이터 변환)
- [표현식 for 항목 in 리스트]

```python
nums = list(range(0,11))
nums
```

```python
nums_div =[]
for value in nums:
    nums_div.append(value//2)
nums_div    
```

```python
[value//2 for value in nums ]
```

#### Type 2. 필터링형 (조건에 맞는 것만)
- [표현식 for 항목 in 리스트 if 조건]

```python
# 짝수만 남기기
evens = [n for n in nums if n % 2 == 0]
print(evens) # [2, 4]
```

#### Type 3. 선택형 (조건에 따라 다르게)
- [참값 if 조건 else 거짓값 for 항목 in 리스트]

# 7장 코드의 조립과 재사용 (함수)

- 함수(Function)는 "입력을 넣으면 특정 기능을 수행하고 출력을 뱉는 기계"와 같습니다.
- 똑같은 코드를 여러 번 복사/붙여넣기 하지 않고, 이름만 불러서 재사용하기 위해 만듭니다.
- **키워드:** `def` (define, 정의하다)
- **매개변수:** 함수라는 기계에 넣는 **재료(입력**)입니다.
- **반환값(`return`):** 기계가 작업을 마치고 뱉어내는 **결과물(출력)**입니다. `return`을 만나면 함수는 즉시 종료됩니다.

```python
# 함수정의, 순서대로 상응하는 매개변수에 복사하는 위치 인수
def add(a, b):         # a, b는 재료(매개변수)
    result = a + b
    return result      # 결과물을 밖으로 던져줌

total = add(3, 5)      # 3과 5를 넣어서 나온 결과(8)를 받음
print(total)
```

# 8장 효율적인 설계도 (클래스와 객체)
- 지금까지는 변수와 함수를 따로 관리했습니다.
- 하지만 게임 캐릭터나 AI 모델처럼 **"속성(변수)"과 "행동(함수)"을 하나로 묶어 관리**해야 할 때가 많습니다. 
- 이것이 바로 **객체지향 프로그래밍(OOP)**입니다.
- 가장 유명한 비유인 **붕어빵 틀**을 기억하세요.
- **클래스 (Class):** 붕어빵 **틀** (설계도)
- **객체/인스턴스 (Object/Instance):** 틀로 찍어낸 **붕어빵** (실체)

```python
class FishBread:
    # 생성자: 붕어빵이 태어날 때 실행됨
    def __init__(self, flavor, price):
        self.flavor = flavor # 맛 (속성)
        self.price = price   # 가격 (속성)

    # 메서드: 붕어빵의 행동
    def show_info(self):
        # 여기서 self는 호출한 녀석(bread1 또는 bread2)이 됩니다.
        print(f"{self.flavor} 붕어빵은 {self.price}원입니다.")

# 객체 생성 (틀로 빵 찍어내기)
bread1 = FishBread("팥", 1000)      # self가 bread1이 되어 '팥'을 저장
bread2 = FishBread("슈크림", 1200)  # self가 bread2가 되어 '슈크림'을 저장

bread1.show_info() # bread1의 정보 출력
bread2.show_info() # bread2의 정보 출력
```

#### **객체 생성 과정의 비밀 (4단계**)
`bread1 = FishBread("팥", 1000)` 코드가 실행될 때, 컴퓨터 내부에서는 다음과 같은 일이 일어납니다.
1. **빈 붕어빵 탄생:** 파이썬이 메모리에 **아무 내용 없는 빈 객체**를 하나 만듭니다.
2. **`self`로 전달:** 이 빈 객체를 `__init__` 함수의 첫 번째 파라미터인 `self`에게 쏙 넣어줍니다. (즉, 지금부터 `self`는 방금 만든 그 빈 붕어빵입니다.)
3. **속성 부여:** `__init__` 함수 안의 `self.flavor = flavor` 코드가 실행되면서, 빈 붕어빵(`self`)에 '팥'이라는 내용물이 채워집니다. 
4. **변수에 저장:** 내용물이 다 채워진 완성된 붕어빵이 `bread1`이라는 변수에 담깁니다.

#### **메서드에서 `self`가 변하는 마법**
객체를 만들고 나서 `bread1.show_info()`를 실행하면 무슨 일이 일어날까요?
- 우리는 `show_info()` 괄호 안에 아무것도 안 넣었지만, 파이썬이 알아서 **주인공**을 넣어줍니다.  
- `bread1.show_info()` ➡ 실제로는 `FishBread.show_info(bread1)`처럼 동작합니다.
- 함수 정의의 `def show_info(self):`에서 `self` 자리에 `bread1`이 들어가는 것이죠.
- 그래서 `print(self.flavor)`는 `print(bread1.flavor)`가 되어 "팥"이 출력되는 것입니다.

### **문제:** `Car` 클래스를 만드세요.
1. 속성: `model`(모델명), `speed`(속도, 초기값 0)
2. 메서드 `accelerate(value)`: 속도가 value만큼 증가
3. 메서드 `brake(value)`: 속도가 value만큼 감소 (0 밑으로는 안 내려감)
4. 차를 한 대 뽑아서 달리고 멈춰보세요.

```python
class Car:
    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self, value):
        self.speed += value
        print(f"{self.model} 가속! 현재 속도: {self.speed}km/h")

    def brake(self, value):
        self.speed -= value
        if self.speed < 0:
            self.speed = 0
        print(f"{self.model} 감속! 현재 속도: {self.speed}km/h")

my_car = Car("테슬라")
my_car.accelerate(50)
my_car.brake(20)
```
